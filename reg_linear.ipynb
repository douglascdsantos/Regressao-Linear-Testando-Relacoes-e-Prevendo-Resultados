```python
url = 'https://raw.githubusercontent.com/douglascdsantos/Regressao-Linear-Testando-Relacoes-e-Prevendo-Resultados/main/Consumo_cerveja.csv'
dados = pd.read_csv(url, sep = ';')
dados.head()
```

## Comportamento da variável dependente
- [x] gráfico de série temporal
- [x] colocar labels e titulo no gráfico

```python
fig, ax = plt.subplots(figsize = (15,6))
ax = plt.title('Consumo de Cerveja em 2015')
ax = plt.xlabel('Dia')
ax = plt.ylabel('Consumo (L)')
ax = dados['consumo'].plot()
```
